# Anális de Datos de Fraude en Seguros de Auto

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Nombramos el marco de datos
car_fraud = pd.read_csv('.\..\data\Dataset.csv')

# Mostramos el encabezado de los datos
car_fraud.head()
```

## 1. Análisis Exoloratorio de los Datos

```python
# Mostramos el número de filas y columnas, el tipo de datos y valores nulos
car_fraud.info()
```

```python
# Resumimos las variables numéricas
car_fraud.describe()
```

```python
# Resuminos las variables categóricas
car_fraud.describe(include = ['O'])
```

### 1.1 Análisis de la Variable Objetivo

```python
# Distribución del fraude en el total de reclamaciones
car_fraud.groupby(['FraudFound_P']).count()['PolicyNumber']
```

```python
# Cómo se distribuye el fraude por sexo del asegurado
car_fraud.groupby(['FraudFound_P', 'Sex']).count()['PolicyNumber']
```

```python
grouped_sex = car_fraud.groupby(['FraudFound_P', 'Sex']).count()['PolicyNumber']
grouped_sex.unstack(level=0).plot.bar()
plt.show()
```

```python
# Creamos un dataframe solo para fraude encontrado
fraud_found = car_fraud[car_fraud['FraudFound_P'] == 1]
# Cómo se distribuye el fraude por edad del asegurado
fraud_found['Age'].plot.hist()
plt.ylabel('Frecuency')
plt.xlabel('Age')
```
